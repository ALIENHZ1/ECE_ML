## Problem 2.2

```python

heart_df.loc[heart_df['age'] < 0, 'age'] = np.nan
heart_df.loc[heart_df['age'] > 125, 'age'] = np.nan

heart_df.loc[(heart_df['anaemia'] != 1) & (heart_df['anaemia'] != 0), 'anaemia'] = np.nan

heart_df.loc[heart_df['creatinine_phosphokinase'] < 0, 'creatinine_phosphokinase'] = np.nan
heart_df.loc[heart_df['creatinine_phosphokinase'] > 10000, 'creatinine_phosphokinase'] = np.nan

heart_df.loc[(heart_df['diabetes'] != 1) & (heart_df['diabetes'] != 0), 'diabetes'] = np.nan

heart_df.loc[heart_df['ejection_fraction'] < 0, 'ejection_fraction'] = np.nan
heart_df.loc[heart_df['ejection_fraction'] > 100, 'ejection_fraction'] = np.nan

heart_df.loc[(heart_df['high_blood_pressure'] != 1) & (heart_df['high_blood_pressure'] != 0), 'high_blood_pressure'] = np.nan

heart_df.loc[heart_df['platelets'] < 20000, 'platelets'] = np.nan
heart_df.loc[heart_df['platelets'] > 1000000, 'platelets'] = np.nan

heart_df.loc[heart_df['serum_creatinine'] < 0, 'serum_creatinine'] = np.nan
heart_df.loc[heart_df['serum_creatinine'] > 10, 'serum_creatinine'] = np.nan

heart_df.loc[heart_df['serum_sodium'] < 100, 'serum_sodium'] = np.nan
heart_df.loc[heart_df['serum_sodium'] > 175, 'serum_sodium'] = np.nan

heart_df.loc[(heart_df['sex'] != 1) & (heart_df['sex'] != 0), 'sex'] = np.nan

heart_df.loc[(heart_df['smoking'] != 1) & (heart_df['smoking'] != 0), 'smoking'] = np.nan

heart_df.loc[heart_df['time'] < 0, 'time'] = np.nan
heart_df.loc[heart_df['time'] > 700, 'time'] = np.nan

heart_df.loc[(heart_df['DEATH_EVENT'] != 1) & (heart_df['DEATH_EVENT'] != 0), 'DEATH_EVENT'] = np.nan
```

The oldest age ever recorded is 122 years old, so 125 is a good upper limit. Due to Rhabdomyolysis, it is possible for creatinine phosphokinase to reach up to 10000 mcg/L, so 10000 is a good upper bound. A percent for ejection fraction can never be above 100. Due to thrombocytosis, it is possible for platelet counts to reach up to 1000000 kiloplatelets/mL. With liver dysfunction it's possible to hit high single digits, which makes 10 a good upper limit. Hyponatremia and Hypernatremia suggest that it's possible to reach limits as low as 100 or as high as 175, which makes those good bounds to use for the data. A valid range that I have instated for time between checkups is up to 2 years without a checkup. All lower bounds not specifically stated are considered to be non-negative values.

## Problem 2.3

```python
curr_columns = heart_df.columns
for col in curr_columns:
    if (col == 'anaemia' or col == 'diabetes' or col == 'high_blood_pressure' or col == 'sex' or col == 'smoking' or col == 'DEATH_EVENT'):
        continue
    curr_mean = heart_df[col].mean(skipna=True)
    curr_std = heart_df[col].std(skipna=True)
    heart_df[col] = (heart_df[col]-curr_mean)/curr_std

heart_df.head(20)
```
